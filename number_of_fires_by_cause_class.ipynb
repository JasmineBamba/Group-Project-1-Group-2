```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Number of fires by cause class.csv', delimiter='\t', encoding='utf-16')
```

```python
# Manually set the column names based on the actual data
df.columns = df.iloc[0]
df = df.iloc[1:].reset_index(drop=True)
df.head(10)
```

```python
# Drop the columns for years 1990-2000
#df.drop(columns=df.loc[:, '1990':'2000'].columns, inplace=True)

# Verify the updated DataFrame
#print(df)
```

```python
# Drop rows that contain NaN values in any column
df = df.fillna(0)

# Reset the index after dropping rows
df.reset_index(drop=True, inplace=True)

#Dropping Data Qualifier column
df.drop("Data Qualifier", axis=1, inplace=True)
    
df.head(15)
```

```python
#Group the data by Cause
df_grouped = df.groupby('Cause').sum()

#Drop the "Unspecified" row 
df_grouped = df_grouped[df_grouped.index != 'Unspecified']

#Reset the index to make 'Month' a regular column
df_grouped.reset_index(inplace=True)

df_grouped
```

```python
print(type(df_grouped))
```

```python
# Bar Plot of Total Wildfires by Cause

colors = ['orange', 'green', 'red','purple']

ax = df_grouped.sum(axis=1).plot(kind='bar', color=colors)
plt.xlabel('Cause')
plt.ylabel('Total Wildfires')
plt.title('Total Wildfires by Cause')
plt.xticks(range(len(df_grouped.index)), df_grouped.index, rotation=45)
plt.tight_layout()


#Adding Labels for Cause
Cause = ["Human Activity", "Lightning", "Prescribed burn", "Reburn"]
tick_locations = range(len(Cause))
plt.xticks(tick_locations, Cause)


# Add data labels above the bars
for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

plt.show()
```

```python
# Group the data by jurisdiction
grouped_data = df.groupby('Jurisdiction').sum()

# Print the grouped data
grouped_data
```

```python

# Visualize the total occurrences for each jurisdiction
plt.figure(figsize=(12, 6))
grouped_data['Total Occurrences'].plot(kind='bar', color='skyblue')
plt.xlabel('Jurisdiction')
plt.ylabel('Total Occurrences')
plt.title('Total Occurrences by Jurisdiction')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
By using df.reset_index(inplace=True), we convert the 'Jurisdiction' column back to a regular column, which should resolve the KeyError when grouping the data. The rest of the analysis and visualization should now work correctly.






```
