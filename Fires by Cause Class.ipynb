# Fires by Cause Class Analysis:

### Primary Causes: 

* The data shows that the primary causes of wildfires are human activity and lightning. These two causes contribute significantly to the overall number of wildfires recorded over the years.

### Variability of Human Activity: 

* Human activity as a cause tends to exhibit higher variability. This could be influenced by various factors such as urbanization, population growth, land use changes, and human behavior. These factors may change over time, leading to fluctuations in the number of wildfires caused by human activity.

### Natural Variations in Lightning:

* Lightning, being a natural cause, is also subject to variations based on climate conditions, weather patterns, and other ecological factors. These natural variations can result in different wildfire counts in different years.

### Low Occurrence of Prescribed Burns and Reburns: 

* The data shows that prescribed burns and reburns are relatively infrequent causes of wildfires compared to human activity and lightning. Prescribed burns are planned and intentional fires conducted for land management purposes, while reburns refer to subsequent fires in an area that has already experienced a wildfire.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Number of fires by cause class.csv', delimiter='\t', encoding='utf-16')
```

```python
# Manually set the column names based on the actual data
df.columns = df.iloc[0]
df = df.iloc[1:].reset_index(drop=True)
df.head(10)
```

```python
# Drop rows that contain NaN values in any column
df = df.fillna(0)

# Reset the index after dropping rows
df.reset_index(drop=True, inplace=True)

#Dropping Data Qualifier column
df.drop("Data Qualifier", axis=1, inplace=True)
    
df.head(15)
```

```python
# Step 1: Calculate the total fires by jurisdiction for each year
total_fires_by_jurisdiction = df.groupby(['Jurisdiction', 'Cause']).sum().reset_index()
total_fires_by_jurisdiction
```

```python
#Group the data by Cause
df_grouped = df.groupby('Cause').sum()



#Reset the index to make 'Month' a regular column
df_grouped.reset_index(inplace=True)

df_grouped
```

```python
print(type(df_grouped))
```

```python
# Bar Plot of Total Wildfires by Cause

df_grouped_sorted = df_grouped.sum(axis=1).sort_values(ascending=False)
colors = ['orange', 'green', 'red','purple', 'yellow']

ax = df_grouped.sum(axis=1).plot(kind='bar', color=colors)
plt.xlabel('Cause')
plt.ylabel('Total Wildfires')
plt.title('Total Wildfires by Cause', color='crimson')
plt.xticks(range(len(df_grouped.index)), df_grouped.index, rotation=45)
plt.tight_layout()


#Adding Labels for Cause
Cause = ["Human Activity", "Lightning", "Prescribed reburn", "Reburn", "Unspecified"]
tick_locations = range(len(Cause))
plt.xticks(tick_locations, Cause)


# Add data labels above the bars
for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

# Save the graph as an image file (e.g., PNG format)
plt.savefig('bar_plot_wildfires.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Summing the values for each cause
total_wildfires_by_cause = df_grouped.sum(axis=1)

# Data for the pie chart
cause_labels = df_grouped.index
cause_values = total_wildfires_by_cause.values
colors = ['orange', 'green', 'red', 'purple']

# Create the pie chart
plt.pie(cause_values, labels=cause_labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Total Wildfires by Cause')

# Display the pie chart
plt.axis('equal')
plt.show()
```
