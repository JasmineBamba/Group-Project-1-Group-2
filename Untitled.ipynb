```python
# Replace 'your_dataset.txt' with the actual file path of your downloaded dataset
df = pd.read_csv('Number of fires by month.csv', delimiter='\t', encoding='utf-16')
```

```python
# Manually set the column names based on the actual data
df.columns = df.iloc[0]
df = df.iloc[1:].reset_index(drop=True)
```

```python
df.head(15)
```

```python
# Drop rows that contain NaN values in any column
df = df.fillna(0)

# Reset the index after dropping rows
df.reset_index(drop=True, inplace=True)
```

```python
df.head(15)
```

```python
#Group the data by Month
df_grouped = df.groupby('Month').sum()

#Drop the "Unspecified" row 
df_grouped = df_grouped[df_grouped.index != 'Unspecified']

#Reset the index to make 'Month' a regular column
df_grouped.reset_index(inplace=True)

df_grouped
```

```python
print(type(df_grouped))
```

```python
#Line Chart

df_grouped.set_index('Month').T.plot(legend=True)
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Monthly Data Over the Years')
plt.show()
```

```python
# Heatmap
plt.figure(figsize=(16, 8))
plt.imshow(df_grouped.drop('Month', axis=1), cmap='YlGnBu', aspect='auto')
plt.colorbar()
plt.xticks(range(len(df_grouped.drop('Month', axis=1).columns)), df_grouped.drop('Month', axis=1).columns, rotation=90)
plt.yticks(range(len(df_grouped)), df_grouped['Month'])
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Monthly Data Heatmap')
plt.tight_layout()
plt.show()
```

```python
#Which month has the highest total value over the years?

max_total_value = df_grouped.drop('Month', axis=1).sum(axis=1).max()
max_value_month = df_grouped[df_grouped.drop('Month', axis=1).sum(axis=1) == max_total_value]['Month'].values[0]

print("Month with the highest total value:", max_value_month)
print("Highest Total Value:", max_total_value)
```

```python
#Which month has the lowest total value over the years?

# Convert the 'Month' column to a string data type
df_grouped['Month'] = df_grouped['Month'].astype(str)

min_total_value = df_grouped.drop('Month', axis=1).sum(axis=1).min()
min_value_month = df_grouped[df_grouped.drop('Month', axis=1).sum(axis=1) == min_total_value]['Month'].values[0]

print("Month with the lowest total value:", min_value_month)
print("Lowest Total Value:", min_total_value)
```

```python
#Box Plot

plt.boxplot(df_grouped.drop('Month', axis=1).values, labels=df_grouped.drop('Month', axis=1).columns, vert=False)
plt.xlabel('Value')
plt.title('Monthly Data Box Plot')
plt.tight_layout()
plt.show()
```

```python
# Bar Chart

df_grouped.set_index('Month').plot(kind='bar', stacked=True)
plt.xlabel('Month')
plt.ylabel('Value')
plt.title('Monthly Data Over the Years')
plt.legend(title='Year', bbox_to_anchor=(1, 1))
plt.show()
```
